# Numpy Array Operations:

**1. Create a 1D NumPy Array:**

```python
import numpy as np
temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])
print(temperatures_w1 )
```

**2. Inspection and Properties:**

```python
print(temperatures_w1.shape)         #shape of the temperatures_w1 array.
print(temperatures_w1.dtype)         #data type of the temperatures_w1 array.
print(temperatures_w1.size)          #size of the temperatures_w1 array.
```

**3. Array Operations:**

```python
temp_Fahrenheit=(temperatures_w1 * 9/5) + 32
print(temp_Fahrenheit)
print(np.max(temp_Fahrenheit))                    #maximum temperatures for the week.
print(np.min(temp_Fahrenheit))                    #minmum temperatures for the week.
print(np.mean(temp_Fahrenheit))                   #mean temperatures for the week.
```

**4. Array Slicing and Indexing:**

```python
print(temp_Fahrenheit[0:3])           #temperatures for the first three days of the week.
print(temp_Fahrenheit[5:])            #temperatures for the weekend (last two days).
print( temp_Fahrenheit[2:5])          #temperatures for the middle three days of the week.
```

**5. Create a 2D Array:**

```python
temperatures = np.array([[22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9], [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1] ])
print(temperatures)
```

**6. Inspect and Slice the 2D Array:**

```python
print(temperatures.shape)         #shape of the temperatures array.
print(temperatures.dtype)         #data type of the temperatures array.
print(temperatures.size)          #size of the temperatures array.
print(temperatures[0])            #temperatures of week 1.
print(temperatures[1])            #temperatures of week 2.
print(temperatures[0,5:])         #weekend temperatures of week 1.
print(temperatures[1,5:])         #weekend teperatures of week 2.
```

# Pandas Series:

### **1. Creating Pandas Series:**

```python
import pandas as pd
marks = [ 95, 92, 89, 85, 80 ]
custom_indices = [ 'Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5' ]
marks = pd.Series(marks, index = custom_indices)
marks
```

### **2. Indexing and Slicing:**

```python
marks.iloc[0]                    # integer index position to access the mark of the 1st rank student.
```

```python
marks.loc["Rank1":"Rank3"]      #loc accessor to retrieve the marks of the top 3 ranks.
```

```python
marks.iloc[2]                     #iloc accessor to retrieve the mark of the 3rd rank student.
```

```python
marks[marks>90]         # boolean mask to find & retrive ranks where the marks are greater than 90.
```

### **3. Manipulating Series:**

```python
marks.loc['Rank1'] = 100         #Modify the mark of the student with the 1st rank to 100.
marks
```

```python
marks.drop('Rank5')        #Remove the entry corresponding to the last rank from the Series.
```

```python
CGPA = marks / 10            # CGPA by dividing each mark in the Series by 10.
CGPA
```

## **Pandas DataFrame:**

### **1. Creating Pandas DataFrame:**

```python
import pandas as pd
transactions = {
   'TransactionID'   : [ 101, 102, 103, 104, 105, 106, 107, 108, 109, 110 ],
   'ProductCategory' : ['Electronics', 'Clothing', 'Electronics', 'Furniture','Clothing',
                        'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
         'Region'    : ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West',
                        'South', 'North' ],
        'Amount'     :  [200,150, 300, 450, 200, 250, 300, 180, 350, 400]

}
df = pd.DataFrame(transactions)

df
```

## 2. Data Exploration:

```python
print("Dataframe:\n",df)                   # DataFrame and its basic information
print("\n Head: \n",df.head())             # head
print("\n Tail: \n",df.tail())             # tail
print("\n shape: \n",df.shape)             # shape
print("\n Column Names: \n",df.columns)    # column names
print("\n Data Types: \n",df.dtypes)       # data types
```

```python
df[['ProductCategory','Amount']]            #Display only the 'ProductCategory' and 'Amount' columns.
```

```python
df.iloc[:,-3:]                                        #Retrieve the last 3 columns of the DataFrame using iloc or loc
```

```python
df[(df["Region"]=='North') & (df["Amount"]>200)]                 #Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200.
```

```python
df['ProductCategory'].value_counts()                        #Find the value counts for the 'ProductCategory' column.
```

```python
df['Region'].unique()                                             #Find the unique values in the 'Region' column.
```

```python
df.groupby('Region')['Amount'].mean()                            #Group by 'Region' and find the mean amount for each region.
```

**3. Manipulating the DataFrame:**

```python
df.loc[df['TransactionID']==102,'Amount']=165      #Modify the 'Amount' for the transaction with TransactionID 102 to 165.
df
```

```python
df['Discount']= df['Amount'] / 10                    # Add a new column 'Discount' by calculating 10% of the 'Amount'.
df
```

```python
df.drop(8)                      #Remove the row with TransactionID 109.
```

```python
df.drop('Discount',axis=1)            #Delete the 'Discount' column from the DataFrame.
```
