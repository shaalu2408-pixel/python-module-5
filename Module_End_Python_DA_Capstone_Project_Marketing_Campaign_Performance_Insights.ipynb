# **3) Exploratory Data Analysis (EDA) and Visualization**

# **Campaign Performance**:

## **Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.**

```python
plt.figure(figsize=(10,5))
plt.scatter(df.Acquisition_Cost,df.ROI,color='lightgreen',edgecolor= 'black',marker='o',alpha=1.0)
plt.title('Relationship between Acquisition_Cost and ROI')
plt.xlabel('Acquisition_cost')
plt.ylabel('ROI')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## **Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.**

```python
avg_con_rate=df.groupby(["Channel_Used","Campaign_Type"])["Conversion_Rate"].mean().reset_index()
avg_con_rate
```

```python
plt.figure(figsize=(10,5))
bars=plt.bar(avg_con_rate.Channel_Used,avg_con_rate.Conversion_Rate,color='purple',edgecolor='black',alpha=1.0)
plt.title('Avg Conversion_Rate for Channel_Used vs Campaign_Type ')
plt.xlabel('Channel_Used')
plt.xticks(rotation=60)
plt.ylabel('Conversion_Rate')
plt.show()
```

## **Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.**


```python
Campaign_Type=df['Campaign_Type'].unique()
data = [df[df['Campaign_Type']==ct]['Engagement_Score']for ct in Campaign_Type]
plt.figure(figsize=(10,5))
plt.boxplot(data,patch_artist=True,
            boxprops=dict(facecolor='lightgreen',color='orange'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o',markerfacecolor='black',markersize=5,alpha=0.8))
plt.title('Distribution of Engagement_Score across diff Campaign_Type')
plt.xlabel('Campaign_Type')
plt.ylabel('Engagement_Score')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

# Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies

```python
avg_roi = df.groupby('Company')['ROI'].mean().reset_index()
avg_roi = avg_roi.sort_values(by='ROI', ascending=False)
avg_roi
```

```python
plt.figure(figsize=(10,5))
bars=plt.bar(avg_roi.Company,avg_roi.ROI,color='skyblue',edgecolor='black',alpha=1.0)
plt.title('Profitability of campaigns conducted by different companies')
plt.xlabel('Company')
plt.xticks(rotation=60)
plt.ylabel('ROI')
for bar in bars:
  height=bar.get_height()
  plt.text(
      bar.get_x()+bar.get_width()/2,height/2,f'{height:.1f}',
      ha='center',va='center',color='black',fontsize=10,fontweight='bold'
  )
plt.show()
```

## **Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.**

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix,annot=True)
plt.xticks(rotation=45)
plt.show()
```

#  Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.
